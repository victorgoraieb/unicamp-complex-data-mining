```
%%file trab2.py

from mrjob.job import MRJob
from mrjob.step import MRStep
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import re
import sys

class Sentiment(MRJob):

  def mapper_init(self):
    nltk.download('vader_lexicon', quiet=True)
    self.sid = SentimentIntensityAnalyzer()

  def mapper(self, _, line):
      
      try:
        tweet_id, author_id, inbound, created_at, text, response_tweet_id, in_response_to_tweet_id = line.split(',')
        polarity = self.sid.polarity_scores(text)

        if inbound == 'False':
          yield author_id, [polarity['compound'], 1]

      except:
        pass

  def combiner(self, key, values) :
      
      acum = 0.0
      counter = 0
      
      for value, freq in values :
          acum += value
          counter += freq
      
      yield key, [acum, counter]

  def reducer(self, key, values):

      acum = 0.0
      counter = 0
      
      for value, freq in values :
          acum += value
          counter += freq
      
      yield key, acum/counter

if __name__ == '__main__':
    Sentiment.run()
```
